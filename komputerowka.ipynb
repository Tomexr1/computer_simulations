```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import scipy.stats as stats
```

```python
def pareto_rvs(a=1, b=1, size=1):
    return b * np.random.uniform(size=size) ** (-1/a)
```

```python
X = pareto_rvs(size=1000)
fig, ax = plt.subplots(3, 1, layout='constrained')
val = np.unique(X)
ax[0].plot(val, stats.pareto.pdf(val,b=1), c='y', label='pdf theoretical')
sns.histplot(X, stat='density', ax=ax[0], color='red', alpha=0.5, label='pdf empirical')
ax[0].set_xlim((0,10))
ax[1].plot(val, stats.pareto.cdf(val, b=1),
         label='cdf theoretical', linestyle=(0, (5, 10)))
sns.ecdfplot(X, label='cdf empirical', color='red', alpha=0.5, ax=ax[1])
ax[1].set_xlim((0,50))
stats.probplot(X, dist='pareto', sparams=(1), plot=ax[2])
fig.legend()
plt.show()
```

## Zad 2

```python
A = 7
T = 5 * 365
r0 = 2 * A
c = A / 3
E_A = A
parisian_limit = 4

# Funkcja generujƒÖca niejednorodny proces Poissona
def inh_poisson_process(lamb, T):
    M = 2
    t, I = 0, 0
    ts = []
    while True:
        Us = np.random.uniform(size=2)
        t -= np.log(Us[0]) / M
        if t > T:
            break
        if Us[1] <= lamb(t)/M:
            I += 1
            ts.append(t)
    return ts

# Funkcja generujƒÖca trajektorie kapita≈Çu
def simulate_capital(T, r0, c, E_A, parisian_limit):
    t = np.arange(T)
    R = np.zeros(T)
    R[0] = r0
    ruin_day = None
    negative_streak = 0

    for i in range(1, T):
        lambda_t = lambda t: 1 + np.sin(t)
        
        events = inh_poisson_process(lambda_t, 1)
        Nt = len(events)
        
        if Nt > 0:
            X = np.random.exponential(E_A, Nt)
        else:
            X = 0
        R[i] = R[i-1] + c - np.sum(X)
        
        if R[i] < 0:
            negative_streak += 1
            if negative_streak > parisian_limit or R[i] < -10:
                ruin_day = i
                break
        else:
            negative_streak = 0
    
    return R, ruin_day

num_simulations = 1000
ruin_days = []

for _ in range(num_simulations):
    R, ruin_day = simulate_capital(T, r0, c, E_A, parisian_limit)
    if ruin_day is not None:
        ruin_days.append(ruin_day)

# Prawdopodobie≈Ñstwo ruiny w ciƒÖgu 5 lat
ruin_prob = len(ruin_days) / num_simulations

# Rozk≈Çad czasu ruiny
plt.hist(ruin_days, bins=50, density=True, alpha=0.6, color='g')
plt.title('Rozk≈Çad czasu ruiny')
plt.xlabel('Dni do ruiny')
plt.ylabel('Gƒôsto≈õƒá prawdopodobie≈Ñstwa')
plt.grid(True)
plt.show()

print(f"Prawdopodobie≈Ñstwo ruiny w ciƒÖgu 5 lat: {ruin_prob}")
```

```python
T = 7
N = 1000  # liczba punkt√≥w czasowych
M = 10  # liczba trajektorii
dt = T / N
t = np.linspace(0, T, N)

W = np.cumsum(np.sqrt(dt) * np.random.randn(M, N), axis=1)

X = W - (t / T) * W[:, -1][:, np.newaxis]
plt.figure(figsize=(12, 8))
for i in range(M):
    plt.plot(t, X[i])
plt.title('Trajektorie mostu Browna')
plt.xlabel('t')
plt.ylabel('$X_t$')
plt.grid(True)
plt.show()
```

```python
E_Xt2 = np.mean(X**2, axis=0)

plt.figure(figsize=(12, 8))
plt.plot(t, E_Xt2)
plt.title('Odchylenie ≈õredniokwadratowe $\mathbb{E}[X_t^2]$')
plt.xlabel('t')
plt.ylabel('$\mathbb{E}[X_t^2]$')
plt.grid(True)
plt.show()
```
